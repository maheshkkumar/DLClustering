```python
#imports 

import pandas as pd
import os
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_style('white')
sns.palplot(sns.color_palette())

%matplotlib inline
```

```python
# loading the results csv files

base_path = './results/mnist/'

files = [os.path.join(base_path, f) for f in os.listdir(base_path)]
```

```python
fig, ax = plt.subplots()

for f in files:
    dataset_lr = f.split(os.sep)[-1].split('.csv')[0].split('_')
    dataset, lr = dataset_lr[0], dataset_lr[1]
    dataframe = pd.read_csv(f)
    ax.plot(dataframe.Iteration, dataframe.Accuracy, label=lr)
    plt.xlabel('Iterations')
    plt.ylabel('Accuracy')
    plt.title('Learning Rates Accuracy (MNIST)')
    ax.legend()

fig.savefig('/home/kumarkm/Dropbox/Courses/Data Mining/ReportTemplate_LaTeX/images/mnist_lr_acc.png', dpi=1000, bbox_inches='tight')
```
